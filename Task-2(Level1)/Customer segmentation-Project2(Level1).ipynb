```python
Q1 = df['MntTotal'].quantile(0.25)
Q3 = df['MntTotal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['MntTotal'] < lower_bound) | (df['MntTotal'] > upper_bound)]
outliers.head()
```

```python
data = df[(df['MntTotal'] > lower_bound) & (df['MntTotal'] < upper_bound)]
data.describe()
```

```python
def get_relationship(row):
    if row['marital_Married']==1:
        return 1
    elif row['marital_Together']==1:
        return 1
    else:
        return 0
df['In relationship']=df.apply(get_relationship, axis=1)
df.head()
```

```python
#correcting a column name
df.rename(columns={'education_2n Cycle\t':'education_2n Cycle'},inplace=True)
df
```

```python
#assign the education status from level 1 to 5 for different rows
def Education_status(row):
    if row['education_2n Cycle']==1:
        return '2'
    elif row['education_Basic']==1:
        return '1'
    elif row['education_Graduation']==1:
        return '3'
    elif row['education_Master']==1:
        return '4'
    elif row['education_PhD']==1:
        return '5'
    else:
        return 'Unknown'
df['Education_status']=df.apply(Education_status,axis=1)
df
    
```

```python
#importing kmeans
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
cols_to_cluster=['Income','MntTotal','marital_Together']
data_scaled=data.copy()
data_scaled[cols_to_cluster] = scaler.fit_transform(data[cols_to_cluster])
```

```python
data_scaled[cols_to_cluster]
```

```python
data_scaled[cols_to_cluster].describe()
```

```python
X=data_scaled[cols_to_cluster]
inertia_list=[]
for i in range(1,11):
    inertia=KMeans(n_clusters=i,random_state=10).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(1,11),inertia_list)
plt.title("Inertia versus No.of clusters")
plt.xlabel("Clusters(i)")
plt.ylabel("Inertia")
plt.show()
```

```python
model=KMeans(n_clusters=3,random_state=8)
model.fit(data_scaled[cols_to_cluster])
data_scaled['Cluster']=model.predict(data_scaled[cols_to_cluster])
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(data_scaled[cols_to_cluster])
data_scaled['pc1'] = pca_res[:,0]
data_scaled['pc2'] = pca_res[:,1]
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='pc1', y='pc2', data=data_scaled, hue='Cluster', palette='magma')
plt.title('Clustered  Visualization')
plt.xlabel('Component1 (pc1)')
plt.ylabel('Component2 (pc2)')
plt.legend(title='Clusters')
```
