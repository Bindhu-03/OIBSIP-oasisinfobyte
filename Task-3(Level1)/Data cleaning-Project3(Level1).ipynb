In the loaded dataset, null values are present in host_name,last_review,reviews_per_month columns.

# cleaning data

```python
#converting dtype and removing null values
df['host_name']= df['host_name'].astype(str)
df['host_name']= df['host_name'].str.replace("nan"," ")


#removing all non-alphabetical characters
df['host_name']=df['host_name'].str.replace('[^a-zA-Z]','',regex=True)

#creating space between two names
df['host_name']=df['host_name'].str.replace(r'([a-z])([A-Z])',r'\1 \2', regex=True)

df.head()
```

```python
#accessing the data where Air bnb is priced
df_priced=df[df['price']!=0]
df_priced
```

```python
#accessing the data where minimum nights is 365 days or less
df_min_nights=df[df['minimum_nights']<=365]
df_min_nights
```

```python
#accessing the data which has atleast an availability for 1day
df_avail=df[df['availability_365']>=1]
df_avail
```

```python
#removing null values from last review column
#converting dtype and removing null values
df['last_review']= df['last_review'].astype(str)
df['last_review']= df['last_review'].str.replace("nan"," ")
df.head()
```

```python
#removing null values from review_per_month column
df['reviews_per_month']= df['reviews_per_month'].astype(str)
df['reviews_per_month']= df['reviews_per_month'].str.replace("nan"," ")
df.head()
```
